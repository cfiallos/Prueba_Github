# **Eliminación de Ruido con Apertura Morfológica**

## ¿Qué es la Apertura Morfológica?

Es una operación de procesamiento de imágenes que permite aplicar una combinación de la erosión y planificación a una imagen, esta operación se utiliza para eliminar el ruido y separar los elementos de una imagen que se encuentran conectados.

## ¿Cúales son las Partes de la Apertura Morfológica?

Sus partes son las siguientes:

- ```Erosión``` => Se encarga de reducir el ruido de la imagen eliminando el ruido de fondo.

- ```Dilatación``` => Se aplica despues de la erosión, expande las áreas significativas para restaurar la estructura de los objetos.

- ```cv2.morphologyEX()``` => Es una función que puede realizar varias operaciones morfológicas en una imagen, algunas de estas operaciones morfológicas serían: apertura, cierre, gradiente, etc.

- ```cv2.MORPH_OPEN``` => Este es un parametro que se utiliza en la función para específicar que se quiere realizar una operación de apertura morfológica.

## Procedimiento detallado de la Erosión y la Dilatación

- La erosión se va a encargar de eliminar los pequeños puntos de ruido aislados, lo que puede hacer que las áreas más grandes de la imagen se reduzcan ligeramente.

- La dilatación se va a encargar de restaurar las áreas significativas de imagen a su tamaño original, pero si el ruido del fundo.

## Ventajas y Desventajas

#### Ventajas

- **Eliminación de Ruido**: Es eficaz para eliminar los pequeños puntos de ruido en una imagen.

- **Preservación de Estructuras**: Conserva las formas y tamaños de los objetos importantes de la imagen.
- **Preparación del Análisis**: Facilita el análisis y el procesamiento posterior de la imagen.

#### Desventajas

- **Perdida de Detalles**: Puede eliminar pequeños detalles importantes junto con el ruido.

- **Sensibilidad al Tamaño del Kernel**: El resultado depende del tamaño y la forma del elemento estructurante utilizado.
- **No Elimina Ruido Complejo**: No es eficaz para eliminar ruidos más complejos o patrones de ruido.

## Aplicaciones Comunes

- Eliminación de ruido en las imágenes médicas.

- Preparación de imágenes para segmentación.

- Mejora de la calidad de imágenes digitales.

## Ejemplo de Aplicación

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen (asegúrate de que la ruta sea correcta)
imagen_original = 'paisaje.png'
image = cv2.imread(imagen_original, cv2.IMREAD_GRAYSCALE)

# Verificar si la imagen se ha cargado correctamente
if image is None:
    raise ValueError("No se pudo cargar la imagen. Verifica la ruta.")

# Definir el elemento estructurante (un disco de radio 5)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))

# Aplicar la apertura morfológica
opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

# Mostrar la imagen original y la imagen procesada
plt.figure(figsize=(10, 10))

# Imagen original
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(image, cmap='gray')
plt.axis('off')

# Imagen después de la apertura
plt.subplot(1, 2, 2)
plt.title('Imagen Después de Apertura')
plt.imshow(opened_image, cmap='gray')
plt.axis('off')

plt.show()
````
