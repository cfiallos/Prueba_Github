# **¿Para qué sirven los "__" al inicio?**

En Python, los nombres de atributos o métodos que comienzan con dos guiones bajos (por ejemplo, "__atributo") utilizan un mecanismo llamado name mangling las clases, especialmente en la herencia.

### ¿Qué es el name mangling?

Cuando defines un atributo o método con dos guiones bajos al inicio, Python procede a modificar su nombre para incluirlo dentro de la clase. 

### ¿Cómo se utiliza?

Por ejemplo, en una clase llamada "MiClase", un atributo definido como __atributo se renombrará internamente a "_MiClase__atributo". 

Esto significa que no se puede acceder a "__atributo"
directamente desde fuera de la clase, pero sí se puede acceder usando el nombre modificado.


### Ejemplo de Aplicación

```py 
class MiClase:
    def __init__(self):
        self.__atributo = 42  
```
Renombrado a _MiClase__atributo
```py 
    def obtener_atributo(self):
        return self.__atributo
```

En conclusión, los dos guiones bajos al inicio de un nombre en python indican el atributo o método es privado y se aplica name mangling para evitar conflictos en la herencia.

# **Kwargs**

### ¿Qué es el Kwargs?

En Python, "**kwargs" es una convención que permite pasar un número variable de argumentos con nombre a una función, convirtiéndose en un diccionario dentro de la función. 

Esto proporciona flexibilidad al capturar argumentos no definidos explícitamente.


### ¿Cómo se utiliza?

El kwargs en el caso de una función, se utiliza para capturar los argumentos adicionales que no se han definido explicitamente en la firma de la función.

También, se convierte en un diccionario donde las claves son los nombres de los argumentos
y los valores, estos vendrían a ser sus respectivos valores.


### Ejemplo Básico de Uso

```py
def funcion_con_kwargs(**kwargs):
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")
funcion_con_kwargs(nombre="Juan", edad=30, ciudad="Madrid")
```

### Uso del Kwargs en Vision por Computadora


Nos permite configurar diversos modelos, de esta manera se puede pasar parámetros como el número de capas y la tasa de aprendizaje al crear modelos de aprendizaje profundo. Por Ejemplo:
```py
def crear_modelo(tipo_modelo, **kwargs):
    # Configuración del modelo según el tipo
    return modelo
```
También, nos facilita el paso de parámetros para técnicas de procesamiento de imágenes. Por Ejemplo:
```py
def preprocesar_imagen(imagen, **kwargs):
    # Aplicar técnicas de preprocesamiento
    return imagen
```
Por último, nos permite ajustar parámetros como el número de épocas y el tamaño del lote de manera flexible.

En conclusión, **kwargs nos permite pasar argumentos con nombre de forma flexible, facilitando la personalización y extensión de funciones.


#   **cv2.blur**

### ¿Qué es el cv2.blur en vision por computadora?

El cv2.blur es una función de la biblioteca OpenCV en Python, que se utiliza para aplicar un filtro de desenfoque promedio a una imagen.
Este filtro resulta muy util en vision por computadora cuando queremos reducir el ruido y los detalles finos en una imagen. 

### ¿Qué es el filtro promedio?

Es un tipo de filtro que ayuda a suavizar la imagen, dando como resultado una imagen menos ruidosa o borrosa.
Esto se realiza mediante el uso de un kernel de tamaño fijo que se desplaza por la imagen, el tamaño de esta ventana determina cuantos pixeles se consideran para calcular el promedio

### ¿Cómo se utiliza cv2.blur?

Para utilizar el cv2.blur es necesario tener instalada la biblioteca de OpenCV, esta la podemos hacer de la siguiente manera:
```py
pip install opencv-python
```
Luego cargamos la imagen que deseamos procesar mediante el uso de OpenCV.
Utilizamos la función cv2.blur para aplicar el filtro a la imagen.
Por último, mostramos la imagen procesada o la guardamos en un archivo. 

````python
import cv2
import matplotlib.pyplot as plt

# Cargamos la imagen que vamos a utilizar
imagen = cv2.imread('cr7.png')

#Convertimos la imagen de BGR a RGB
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)


# Aplicamos el filtro de desenfoque promedio
imagen_desenfocada = cv2.blur(imagen, (30,50))

# Creamos la figura en donde se va a presentar la imagen
plt.figure(figsize=(10, 5))

# Mostramos la imagen original
plt.subplot(1, 2, 1), plt.imshow(imagen), plt.title('Imagen Original')

# Mostramos la imagen desenfocada
plt.subplot(1, 2, 2),plt.imshow(imagen_desenfocada), plt.title('Imagen Desenfocada')

# Mostramos la figura
plt.show()
````

# **El Uso de Sigma en Gaussian Blur**

### ¿Qué es el Sigma en Gaussian Blur?

Es un parametro que representa la desviación estandar de la distribución gaussiana, que es utilizada para aplicar el desenfoque a una imagen.
Este parámetro resulta muy importante ya que determina la cantidad de suavizado que se aplicará a la imagen.

### Ejemplo Funcionamiento del Sigma 

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Cargar una imagen
imagen = cv2.imread('cr7.png')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Aplicar desenfoque gaussiano
sigma = 25
imagen_gaussian_blur = cv2.GaussianBlur(imagen, (5, 5), sigma)

# Mostrar la imagen original y la desenfocada
plt.subplot(1,2,1), plt.imshow(imagen), plt.title('Imagen Original')
plt.subplot(1,2,2), plt.imshow(imagen_gaussian_blur), plt.title('Imagen Gaussian Blur')
````

# **El Valor de K en Harris Corner**

### ¿Qué es y para qué Sirve?

El valor de K en el detector de esquinar Harris Corner, se utiliza en la fórmula para calcular la respuesta de la esquina, este valor es un parámetro que puede influir en la sensibilidad del detector a las esquinas. Comunmente, se recomienda que el valor de K este situado en el rango de 0.04 a 0.06

### Ejemplo de uso 

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = cv2.imread('Cr7.png')

# Convertir a escala de grises
gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Convertir a float32
gray = np.float32(gray)

# Parámetros para el detector de Harris
block_size = 2  # Tamaño de la vecindad
ksize = 3       # Tamaño del filtro Sobel
k = 0.04        # Valor de K

# Aplicar el detector de esquinas de Harris
dst = cv2.cornerHarris(gray, block_size, ksize, k)

# Dilatar la imagen para marcar las esquinas
dst = cv2.dilate(dst, None)

# Umbral para marcar las esquinas
threshold = 0.01 * dst.max()

# Marcar las esquinas en rojo en la imagen original
imagen[dst > threshold] = [0, 0, 255]  # Marcar esquinas en rojo

# Mostrar la imagen con las esquinas detectadas
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))  # Convertir de BGR a RGB para mostrar correctamente
plt.title('Harris Corner')
plt.axis('off')  # Opcional: ocultar los ejes
plt.show()
````
