#   **cv2.blur**

### ¿Qué es el cv2.blur en vision por computadora?

El cv2.blur es una función de la biblioteca OpenCV en Python, que se utiliza para aplicar un filtro de desenfoque promedio a una imagen.
Este filtro resulta muy util en vision por computadora cuando queremos reducir el ruido y los detalles finos en una imagen. 

### ¿Qué es el filtro promedio?

Es un tipo de filtro que ayuda a suavizar la imagen, dando como resultado una imagen menos ruidosa o borrosa.
Esto se realiza mediante el uso de un kernel de tamaño fijo que se desplaza por la imagen, el tamaño de esta ventana determina cuantos pixeles se consideran para calcular el promedio

### ¿Cómo se utiliza cv2.blur?

Para utilizar el cv2.blur es necesario tener instalada la biblioteca de OpenCV, esta la podemos hacer de la siguiente manera:
```py
pip install opencv-python
```
Luego cargamos la imagen que deseamos procesar mediante el uso de OpenCV.
Utilizamos la función cv2.blur para aplicar el filtro a la imagen.
Por último, mostramos la imagen procesada o la guardamos en un archivo. 

### Ejemplo de aplicación con código

````python
import cv2
import matplotlib.pyplot as plt

# Cargamos la imagen que vamos a utilizar
imagen = cv2.imread('cr7.png')

#Convertimos la imagen de BGR a RGB
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)


# Aplicamos el filtro de desenfoque promedio
imagen_desenfocada = cv2.blur(imagen, (30,50))

# Creamos la figura en donde se va a presentar la imagen
plt.figure(figsize=(10, 5))

# Mostramos la imagen original
plt.subplot(1, 2, 1), plt.imshow(imagen), plt.title('Imagen Original')

# Mostramos la imagen desenfocada
plt.subplot(1, 2, 2),plt.imshow(imagen_desenfocada), plt.title('Imagen Desenfocada')

# Mostramos la figura
plt.show()
````

### Consideraciones a la hora de utilizar cv2.blur

Si elegimos un tamaño más grande del kernel nos dará como resultado un mayor nivel de desenfoque, mientras que si elegimos un tamaño más pequeño este tendrá un efecto más sutil.

Por último, hay que tener en cuenta que el uso del filtro promedio nos ayudará a suavizar la imagen, pero también puede eliminar detalles que son importantes, por esta razon es importante
elegir correctamente el tamaño del kernel dependiendo su aplicación.
