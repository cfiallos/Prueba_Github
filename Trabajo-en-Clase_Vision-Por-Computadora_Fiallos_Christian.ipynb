```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Entrada: Ruta de la imagen 
#Proceso: Cargar la imagen
#Salida: Una imagen 
def cargar_imagen(ruta):
    imagen = cv2.imread(ruta)
    return imagen

#Entrada: Una imagen
#Proceso: Convertir una imagen RGB a una imagen en escala de grises
#Salida: Imagen en escala de grises
def convertir_escala_grises(imagen_cargada):
    img_escala_grises = cv2.cvtColor(imagen_cargada, cv2.COLOR_RGB2GRAY)
    return img_escala_grises

#Entrada: Imagen en escala de grises
#Proceso: Suavizar imagen utilizando un filtro
#Salida: Imagen con filtro suavizado
def suavizar_imagen(imagen_escala_grises, size):
    imagen_suavizada = cv2.blur(imagen_escala_grises,(size, size))
    return imagen_suavizada

#Entrada: Imagen Suavizada
#Proceso: Filtrar usando el filtro Sobel
#Salida: Imagen con filtro sobel
def resaltar_bordes(imagen_suavizada, size):
    imagen_bordes_resaltados_x = cv2.Sobel(imagen_suavizada, cv2.CV_64F, 1, 0, (size,size))
    imagen_bordes_resaltados_y = cv2.Sobel(imagen_suavizada, cv2.CV_64F , 0, 1, (size,size))
    sobel = cv2.magnitude(imagen_bordes_resaltados_x, imagen_bordes_resaltados_y)
    return sobel

#Entrada: Imagen bordes resaltados
#Proceso: Filtrar usando el filtro Canny
#Salida: Imagen con filtro Canny
def resaltar_bordes_canny(imagen_bordes_resaltados):
    imagen_bordes_resaltados_canny = cv2.Canny(imagen_bordes_resaltados, -1, 100,200)
    return imagen_bordes_resaltados_canny

imagen_cargada = cargar_imagen("cr7.png")
imagen_escala_grises = convertir_escala_grises(imagen_cargada)
imagen_suavizada = suavizar_imagen(imagen_escala_grises, 8) 
imagen_bordes_resaltados = resaltar_bordes(imagen_suavizada, 8)
imagen_bordes_canny = resaltar_bordes_canny(imagen_escala_grises)


#Presentacion ordenada de imagenes 
plt.subplot(1, 5, 1), plt.imshow(imagen_cargada), plt.title('Imagen Original')
plt.subplot(1, 5, 2), plt.imshow(imagen_escala_grises), plt.title('Escala de Grises')
plt.subplot(1, 5, 3), plt.imshow(imagen_suavizada), plt.title('Filtro Suavizado')
plt.subplot(1, 5, 4), plt.imshow(imagen_bordes_resaltados), plt.title('Filtro Sobel')
plt.subplot(1, 5, 5), plt.imshow(imagen_bordes_canny), plt.title('Imagen Filtro Canny')

#Ruedita Brillo y Contraste




 
```
