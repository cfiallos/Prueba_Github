# Tema 1: Introducción a la Visión por Computadora

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def cargar_imagen(ruta):
    imagen = cv2.imread(ruta)
    return imagen

def mostrar_imagen(imagen, titulo="Imagen"):
    plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
    plt.title(titulo)
    plt.axis('off')
    plt.show()

def convertir_a_grises(imagen):
    return cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

def ajustar_brillo_contraste(imagen, alfa=1.2, beta=50):
    return cv2.convertScaleAbs(imagen, alpha=alfa, beta=beta)

def aplicar_filtro_gaussiano(imagen, kernel_size=5):
    return cv2.GaussianBlur(imagen, (kernel_size, kernel_size), 0)


```

```python
# Prueba del programa
ruta_imagen = '../../../data/raw/lena.png' 
```

```python
imagen_original = cargar_imagen(ruta_imagen)
mostrar_imagen(imagen_original, "Imagen Original")
```

```python
imagen_gris = convertir_a_grises(imagen_original)
mostrar_imagen(imagen_gris, "Imagen en Escala de Grises")
```

```python
imagen_brillo = ajustar_brillo_contraste(imagen_original)
mostrar_imagen(imagen_brillo, "Imagen con Brillo Ajustado")
```

```python
imagen_brillo = ajustar_brillo_contraste(imagen_original, alfa=1.5, beta=100)
mostrar_imagen(imagen_brillo, "Imagen con Brillo Ajustado")
```

```python
imagen_filtrada = aplicar_filtro_gaussiano(imagen_original)
mostrar_imagen(imagen_filtrada, "Imagen con Filtro Gaussiano")
# Dismnuye el ruido de la imagen
```

```python
import cv2
import numpy as np
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

imagen_original = cv2.imread('lena.png')  
imagen_original = cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB)  


def actualizar_imagen(brillo, contraste):
    alpha = contraste / 50  
    beta = brillo - 50  

    imagen_ajustada = cv2.convertScaleAbs(imagen_original, alpha=alpha, beta=beta)
    
    plt.figure(figsize=(6, 4))
    plt.imshow(imagen_ajustada)
    plt.axis('off')
    plt.title(f"Brillo: {brillo}, Contraste: {contraste}")
    plt.show()

brillo_slider = widgets.IntSlider(value=50, min=0, max=100, step=1, description='Brillo')
contraste_slider = widgets.IntSlider(value=50, min=0, max=100, step=1, description='Contraste')

widgets.interactive(actualizar_imagen, brillo=brillo_slider, contraste=contraste_slider)
```

## Conversión de RGB a HSV
El espacio de color **HSV** (Hue, Saturation, Value) es una representación más intuitiva para la visión por computadora, ya que separa la información de color de la intensidad de luz. Se usa en tareas como segmentación de objetos y detección de colores.

### **Por qué se usa HSV?**
- **Separación de la intensidad y color**: La luminancia no afecta el tono del color, facilitando la segmentación.
- **Menor sensibilidad a cambios de iluminación**: El canal de tono (`H`) no se ve afectado por variaciones en el brillo.
- **Útil para detección de colores específicos**: Ideal para aplicaciones de seguimiento de objetos y clasificación de imágenes.

### **Aplicaciones de HSV**
- **Segmentación de objetos por color**: Se puede aislar colores específicos en una imagen.
- **Mejora en detección de bordes**: Evita problemas de iluminación.
- **Análisis en imágenes médicas**: Separa información cromática útil para estudios clínicos.

### **Código para convertir de RGB a HSV**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagen_rgb = cv2.imread('../../../data/raw/lena.png')
imagen_bgr = cv2.cvtColor(imagen_rgb, cv2.COLOR_BGR2RGB)  


imagen_hsv = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2HSV)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1), plt.imshow(imagen_rgb), plt.title('Imagen RGB')
plt.subplot(1,2,2), plt.imshow(imagen_hsv[:,:,0], cmap='hsv'), plt.title('Canal Hue (H)')
plt.show()
```
